# Arithmetic Series

## Modified from Arithmetic Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/ArithmeticSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Arithmetic_progression)


An arithmetic sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous one by adding (or subtracting) a constant $d$.

\begin{align*}
a_1 & \\
a_2 &\ \ =\ \ a_1 + d  \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d   \\
a_4 &\ \ =\ \ a_3 + d \ \ =\ \ a_2 + 2d \ \ =\ \ a_1 + 3d  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ =\  \ a_1+(k-1)d
\end{align*}

Therefore, for $k>=1$,

$$ a_k = a_1 + (k-1)d$$

With $a_1$ and $d$ given.

<br />

The sequence of the partial sum of the first $n$ terms of a sequence is called a series, in the case of an arithemetic sequence the series is called an arithmetic series.

If  $a_1= 1$ and $d=1$ we have, the simplest arithmnetic sequence,  the natural numbers sequence $\{a_k = k \}$ for  $k=1, 2, ..., $  i.e. the seuqnece
$$1, 2, ...,$$
The  corresponding series is
$$
S_n = 1 + 2+ \ldots + (n-1)+ n = \sum_{k=1}^n k
$$
can be calculated as a function of $n$.

Assume that we allign one $S_{n-1}$ with a second  $S_{n-1}$ that have the terms adden reverse order and you add term to term, we have
\begin{array}{llclclclclc}
  S_{n-1} &=& 1 &+& 2 &+& \ldots &+& (n-2) &+& (n-1) \\
  S_{n-1} &=& (n-1)  &+& (n-2)  &+& \ldots   &+& 2   &+& 1 \\
   \hline \\
   2S_{n-1} &=& n  &+& n  &+& \ldots   &+& n   &+& n \\
\end{array}
so
$$  2S_{n-1} = (n-1)n $$
then
$$  S_{n-1} = \frac{1}{2} (n-1)n$$
therefore
$$  S_{n} = \frac{1}{2} n(n+1)$$


\begin{align*}
S_n &= \sum_{k=1}^n a_k	\\
    &= \sum_{k=1}^n[a_1+(k-1)d]		\\
    &= na_1 + d \sum_{k=1}^n(k-1)	\\
    &= na_1 + d \sum_{k=0}^{n-1} k	\\
    &= na_1 + d \sum_{k=1}^{n-1} k	\\
    &= na_1 +  \frac{1}{2} d(n-1)n\\
\end{align*}
so
$$ S_n =  \frac{1}{2} n[2 a_1 +d(n-1)].$$

Note, however, that
$$a_1+a_n=a_1+[a_1+d(n-1)]=2a_1+d(n-1),$$
so
$$S_n = \frac{1}{2}n(a_1+a_n),$$


or $n$ times the arithmetic mean of the first and last terms! This is the trick Gauss used as a schoolboy to solve the problem of summing the integers from $1$ to $100$ given as busy-work by his teacher. While his classmates toiled away doing the addition longhand, Gauss wrote a single number, the correct answer

$$\frac{1}{2}(100)(1+100)=50Â·101=5050$$
on his slate (Burton 1989, pp. 80-81; Hoffman 1998, p. 207). When the answers were examined, Gauss's proved to be the only correct one.

<br />

If  the sequence starts at index $0$, $\{a_k\}$ for $ k=0, 1, 2, ..., $ we have

\begin{align*}
a_0 & \\
a_1 &\ \ =\ \ a_0 + d  \\
a_2 &\ \ =\ \ a_1 + d \ \ =\ \ a_0 + 2d   \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d \ \ =\ \ a_0 + 3d  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ = \  \ a_1+(k-1)d \ \ =\ \ a_0+kd
\end{align*}

<br />
for given   $a_0$ and $d$.

In this case the sum of the terms between $0$ and $n$ of  the arithmetic sequence  is then given by
\begin{align*}
S_n &= \sum_{k=0}^{n} a_k	\\
    &= \sum_{k=0}^{n}[a_0+kd]		\\
    &= (n+1) a_0 + d \sum_{k=0}^{n} k	\\
    &= (n+1) a_0 + d \sum_{k=1}^{} k	\\
    &=(n+1) a_0+  \frac{1}{2} dn(n+1) \\
\end{align*}
so
$$ S_n =  \frac{1}{2} (n+1)[2 a_0 + dn].$$
<br />
$$ S_n =  \frac{1}{2} (n+1)[a_0+ a_n].$$

# Geometric Series

## Modified from Geometric Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/GeometricSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Geometric_series)


An geometric series is the sum of a sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous  by multipliying constant $r$ and starts with 1.

\begin{align*}
a_0 &\ \ =\ \ r^0 \ \ =\ \ 1\\
a_1 &\ \ =\ \ a_0 *  r \ \ =\ \  r\\
a_2 &\ \ =\ \ a_1 * r\ \ =\ \ a_0 * r^2  \ \ =\ \ r^2 \\
a_3 &\ \ =\ \ a_2 * r \ \ =\ \ a_0 * r^3 \ \ =\ \  r^3  \\
a_4 &\ \ =\ \ a_3 * r\ \  =\ \ a_0 * r^4 \ \  =\ \ r^4  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}* r \ =\   a_0 * r^k \ \  =\ \  r^k
\end{align*}

The sum of the first $n$ terms  of  the geometric sequence (called an geometric serie)  is then given by
$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  (1 + r + r^2 + \cdots + r^{n-1} ) +r^n \\
 &= S_{n-1}  +r^n
\end{array}
$$

$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  1 + (r + r^2 + \cdots + r^{n-1}  +r^n)  \\
 &=   1 + r (1 +  r  + r^2 + \cdots + r^{n-1} )  \\
  &=   1 + r   S_{n-1}
\end{array}
$$

Then
$$
 S_{n-1}  +r^n =   1 + r   S_{n-1}
$$

Therefore
$$
 S_{n-1} (1 -r)   +r^n =   1 - r^n
$$

$$
 S_{n-1}   =   \frac{1 - r^n }{1 -r}
$$

$$
 S_{n}   =   \frac{1 - r^{n+1} }{1 -r}  = \frac{ r^{n+1} - 1}{r -1}
$$


## Nodes in Complete Binary Tree with $L=0,1,2,\cdots , L-1$ levels

![](https://raw.githubusercontent.com/gjhernandezp/algorithms/master/Lab7/ComBinTree.png)

In a complete binary tree you have in ecah level $k =0,1,2,\cdots $  $ 2^k $

$$
\begin{array}{ll}
\text {Level } 0 \rightarrow 2^0 = 1 \text { node (root)} \\
\text {Level } 1 \rightarrow 2^1 = 2 \text { nodes} \\
\text {Level } 2 \rightarrow 2^2 = 4 \text { nodes} \\
\text {Level } 3 \rightarrow 2^3 = 8 \text { nodes} \\
\vdots\\
\text {Level } k \rightarrow 2^k  \text { nodes} \\
\end{array}
$$

nodes, with $l = 0,1,2, ... $ the level

Total number of the complte tree with $L=0,1,2,\cdots , L-1$ levels

$$
2^0 + 2^1 + 2^2 + \cdots +2^{L-1}  = 1 + 2 + 4 + \cdots +2^{L-1} = \sum_{k=0}^{L-1} 2^k = \frac{ 2^{L} - 1}{2 -1} = 2^{L} - 1
$$

## Add the a geometric series with $r =2$ up to 10** 6 and verify with the formula.  

```
n = 1000
r = 2

# TÃ©rminos y sumas parciales
terms = [r ** i for i in range(n + 1)]
partial_sums = [sum(terms[:i + 1]) for i in range(n + 1)]

# Suma bucle
sum_n = partial_sums[-1]

# Suma fÃ³rmula
sum_formula = (r ** (n + 1) - 1) / (r - 1)

print(f"Suma bucle: {sum_n}")
print(f"Suma fÃ³rmula: {sum_formula}")
print(f"Diferencia: {abs(sum_n - sum_formula)}")
```

## Your own examples of a geometric series with $r=\frac{1}{d+1}$  with $d$ add the series up to 10**3 and verify with the formula.

```


import matplotlib.pyplot as plt

def geometric_sum(d, n=10**3):
    r = 1 + d
    sequence_an = [r**i for i in range(n+1)]
    sequence_sn = [sum(sequence_an[:i+1]) for i in range(n+1)]

    # geometric series formula
    sum_formula = (1 - r**(n+1)) / (1 - r)

    #  loop
    sum_n = sum(sequence_an)

    return sequence_an, sequence_sn, sum_n, sum_formula


d = 0.1
sequence_an, sequence_sn, sum_n, sum_formula = geometric_sum(d)

# results
print(f"Sum using loop for d = {d}: {sum_n}")
print(f"Sum using formula for d = {d}: {sum_formula}")
print(f"Difference: {abs(sum_n - sum_formula)}")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(sequence_sn, color='black', label='Partial Sums')
plt.plot(sequence_an, color='blue', label='Terms')
plt.legend()
plt.grid(True)
plt.title(f'Geometric Series for d = {d}: First 1000 Terms and Partial Sums')
plt.xlabel('n')
plt.ylabel('Value')


plt.show()
```

## Your own example of power sum with $p =$ last digit in the list $+ 1$, i.e., the sum $1^p+2^p+\cdots + n^p$  series up to  $n=10^6$  and verify with the formula.

```

import time

def power_sum(n, p):
    # Calculate the sum of powers: 1^p + 2^p + ... + n^p
    return sum(i**p for i in range(1, n+1))


n = 10**6
last_digit = int(str(n)[-1])
p = last_digit + 1

start_time = time.time()


sum_n = power_sum(n, p)


end_time = time.time()

# result
print(f"Power sum for n = {n} and p = {p}: {sum_n}")
print(f"Time taken: {end_time - start_time} seconds")
```
